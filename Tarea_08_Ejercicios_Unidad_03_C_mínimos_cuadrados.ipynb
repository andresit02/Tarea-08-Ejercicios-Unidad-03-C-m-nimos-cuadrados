1. Dados los datos:

|$x_i$ | 4.0  | 4.2  | 4.5  | 4.7  | 5.1  | 5.5  | 5.9  | 6.3  | 6.8  | 7.1  |
|---------|------|------|------|------|------|------|------|------|------|------|
 $\(y$\) | 102.56 | 130.11 | 113.18 | 142.05 | 167.53 | 195.14 | 224.87 | 256.73 | 299.50 | 326.72 |

a. Construye el polinomio por mínimos cuadrados de grado 1 y calcula el error.
- b. Construye el polinomio por mínimos cuadrados de grado 2 y calcula el error.
- c. Construye el polinomio por mínimos cuadrados de grado 3 y calcula el error.
- d. Construye el polinomio por mínimos cuadrados de la forma 𝑏𝑒^(ax) y calcula el error.
- e. Construye el polinomio por mínimos cuadrados de la forma 𝑏𝑥^ay calcula el error.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos proporcionados
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Ajuste lineal (grado 1)
coef_lineal = np.polyfit(x, y, 1)
pol_lineal = np.poly1d(coef_lineal)
y_lineal = pol_lineal(x)
error_lineal = np.sum((y - y_lineal) ** 2)

# Ajuste cuadrático (grado 2)
coef_cuadratico = np.polyfit(x, y, 2)
pol_cuadratico = np.poly1d(coef_cuadratico)
y_cuadratico = pol_cuadratico(x)
error_cuadratico = np.sum((y - y_cuadratico) ** 2)

# Ajuste cúbico (grado 3)
coef_cubico = np.polyfit(x, y, 3)
pol_cubico = np.poly1d(coef_cubico)
y_cubico = pol_cubico(x)
error_cubico = np.sum((y - y_cubico) ** 2)

# Función exponencial
def func_exp(x, a, b):
    return b * np.exp(a * x)

popt_exp, _ = curve_fit(func_exp, x, y)
y_exp = func_exp(x, *popt_exp)
error_exp = np.sum((y - y_exp) ** 2)

# Función potencia
def func_pot(x, a, b):
    return b * x ** a

popt_pot, _ = curve_fit(func_pot, x, y)
y_pot = func_pot(x, *popt_pot)
error_pot = np.sum((y - y_pot) ** 2)

# Mostrar los resultados
print(f"Polinomio de grado 1: coeficientes = {coef_lineal}, error = {error_lineal}")
print(f"Polinomio de grado 2: coeficientes = {coef_cuadratico}, error = {error_cuadratico}")
print(f"Polinomio de grado 3: coeficientes = {coef_cubico}, error = {error_cubico}")
print(f"Función exponencial: coeficientes = {popt_exp}, error = {error_exp}")
print(f"Función potencia: coeficientes = {popt_pot}, error = {error_pot}")

# Graficar los datos y las curvas ajustadas
plt.scatter(x, y, label='Datos', color='black')
plt.plot(x, y_lineal, label='Grado 1', color='blue')
plt.plot(x, y_cuadratico, label='Grado 2', color='red')
plt.plot(x, y_cubico, label='Grado 3', color='green')
plt.plot(x, y_exp, label='Exponencial', color='orange')
plt.plot(x, y_pot, label='Potencia', color='purple')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste por Mínimos Cuadrados')
plt.show()
```

4. El siguiente conjunto de datos, presentado al Subcomité Antimonopolio del Senado, muestra las
características comparativas de supervivencia durante un choque de automóviles de diferentes clases.
Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de
vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria).

| Tipo                        | Peso promedio (lb) | Porcentaje de presentación |
|-----------------------------|---------------------|----------------------------|
| Regular lujoso doméstico    | 4800                | 3.1                        |
| Regular intermediario doméstico | 3700            | 4.0                        |
| Regular económico doméstico | 3400                | 5.2                        |
| Compacto doméstico          | 2800                | 6.4                        |
| Compacto extranjero         | 1900                | 9.6                        |


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
pesos = np.array([4800, 3700, 3400, 2800, 1900])
porcentajes = np.array([3.1, 4.0, 5.2, 6.4, 9.6])

# Calcular medias
x_mean = np.mean(pesos)
y_mean = np.mean(porcentajes)

# Calcular Sxx y Sxy
Sxx = np.sum((pesos - x_mean)**2)
Sxy = np.sum((pesos - x_mean) * (porcentajes - y_mean))

# Calcular pendiente m y el intercepto b
m = Sxy / Sxx
b = y_mean - m * x_mean

# Imprimir los coeficientes
print(f"Pendiente m: {m}")
print(f"Intercepto b: {b}")

# Preparar la recta de regresión
x_vals = np.linspace(np.min(pesos), np.max(pesos), 100)
y_vals = m * x_vals + b

# Graficar los datos y la recta de regresión
plt.scatter(pesos, porcentajes, label='Datos')
plt.plot(x_vals, y_vals, color='red', label='Recta de regresión')
plt.xlabel('Peso promedio (lb)')
plt.ylabel('Porcentaje de presentación')
plt.title('Regresión lineal por mínimos cuadrados')
plt.legend()
plt.grid(True)
plt.show()
```
